Data Preprocessing

```python
import pandas as pd


data = pd.read_csv('adult.csv')

data = data.drop_duplicates()

data_categorical = data.select_dtypes(exclude='number')

cols = data_categorical.columns.to_list

data = pd.get_dummies(data, columns=['workclass', 'education', 'marital.status', 'occupation',
       'relationship', 'race', 'sex', 'native.country', 'income'], prefix='encoded').astype(int)

data = data.fillna(0)

```
